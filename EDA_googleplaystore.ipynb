```python
def size_process(item):
    if str(item)[-1]=="M":
      res=float(str(item).replace('M',''))
      res=res*1024
      return res
    elif str(item)[-1]=='K':
      res=float(str(item).replace('M',''))
      return res
    else:
      return str(np.nan)

```

```python
print(size_process('19M'))
```

```python
data_copy['Size']=data_copy['Size'].apply(size_process)
```
